```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plotter(filename:str):
    f = open('FiveMinutes.txt','r')
    lines = np.asarray(f.readlines())
    f.close()
    time = []
    temp = []
    for i in lines:
        t,T,stamp = i.split('\t')
        try:
            time.append(float(t))
            temp.append(float(T))
        except:
            continue
    np.asarray(time)
    np.asarray(temp)
    plt.plot(time,temp)
```
